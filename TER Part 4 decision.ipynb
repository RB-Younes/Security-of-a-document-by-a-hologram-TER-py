```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
from google.colab.patches import cv2_imshow
import tensorflow as tf
patch_size = (80,71) #[1, 2, 4, 8, 71, 142, 284, 568]
```



*    **Prediction**



```python
def divide_image_into_patches(image_path, patch_size):
    image =  cv2.imread(image_path)
    height, width = image.shape[:2]
    patch_height, patch_width = patch_size
    patches = []
    cpt = 0
    for y in range(0, height, patch_height):
        for x in range(0, width, patch_width):
            patch = image[y:y+patch_height, x:x+patch_width]
            patches.append(patch)
            cpt+=1
    return patches,cpt
```

```python
def get_patches_mask(patches_mask):
  cpt_R = 0
  cpt_G = 0
  for i,patch in enumerate(patches_mask):
    first_pixel = patch[0, 0]
    if (first_pixel == [0, 255, 0]).all():
     cpt_G += 1
    elif (first_pixel == [0, 0, 255]).all():
     cpt_R += 1
  return cpt_R , cpt_G
```

```python
def calculate_ratios(cpt_R, cpt_G):
    return cpt_R / 40, cpt_G / 40
```



*   final best model


```python
dossier_H = "/content/drive/MyDrive/TER Data/final best color maps/Holo"
dossier_NH = "/content/drive/MyDrive/TER Data/final best color maps/No-Holo"
```

```python
min_G_ratio = 0.3
prediction = []
actual_calss = []
for path in [dossier_H,dossier_NH]:
  for item in os.listdir(path):
          if os.path.isfile(os.path.join(path, item)) :
              print(item)
              patches,cpt = divide_image_into_patches(str(os.path.join(path, item)), patch_size)
              cpt_R , cpt_G = get_patches_mask(patches)
              if path ==  dossier_H :
                G_ratio = cpt_G/40
                print("class = Holo with RED_ratio = {} and GREEN_ratio = {}".format(cpt_R/40 ,cpt_G/40))
                actual_calss.append("Holo")
                if G_ratio >= min_G_ratio :
                    prediction.append("Holo")
                else:
                    prediction.append("No-Holo")
              else:
                G_ratio = cpt_G/40
                print("class = No-Holo with RED_ratio = {} and GREEN_ratio = {}".format(cpt_R/40 ,cpt_G/40))
                actual_calss.append("No-Holo")
                if G_ratio >= min_G_ratio :
                    prediction.append("Holo")
                else:
                    prediction.append("No-Holo")
```

```python
# Initialize variables to count correct predictions
correct_predictions = 0

# Iterate through both lists simultaneously
for pred, actual in zip(prediction, actual_calss):
    if pred == actual:
        correct_predictions += 1

# Calculate accuracy
accuracy = correct_predictions / len(prediction) * 100

print("Accuracy:", accuracy, "%")
```
